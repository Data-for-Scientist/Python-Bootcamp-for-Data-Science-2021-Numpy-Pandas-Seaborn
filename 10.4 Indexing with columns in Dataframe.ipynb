## 10.4  Indexing with columns in Dataframe

First we import numpy and pandas:

```python
import numpy as np
import pandas as pd
```

Then we will read this dataframe from a csv file using the function **pd.read_csv()**:

```python
data1 = pd.read_csv('data/ex5.csv')
data1.head()
```

We can display the labels for all columns using the attribute **columns**:

```python
data1.columns
```

Now we can set the index for this dataframe to be any column using the function **set_index()**, here we choose the column ('year'):

```python
data2 = data1.set_index('year')
data2.head()
```

We can also set the index to be multiple columns:

```python
data3 = data1.set_index(['year','quarters'])
data3
```

To keep the index column in the dataframe as a normal column, we use the argument **drop = False**: 

```python
data1.set_index('year', drop = False)
```

```python
data3
```

We can aslo reset the index back to the original index that is generated by pandas using the function **reset_index()**:

```python
data3.reset_index()
```
