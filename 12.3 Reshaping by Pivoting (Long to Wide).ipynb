## 12.3 Reshaping by Pivoting (Long to Wide)

First we import pandas library:

```python
import pandas as pd
```

As an example, we will read this dataframe from a csv file using the function **pd.read_csv()**:

```python
data1 = pd.read_csv('data/ex9.csv')
data1
```

We can do the opposite of melting which is **pivoting**: i.e. transforming a single column into multiple columns using the function **pivot()**.

Here we transform the column **sales**:

```python
data1.pivot(index = 'year', columns = 'sales', values = 'amount')
```

For another example, we read this dataframe:

```python
data2 = pd.read_csv('data/ex10.csv')
data2
```

We can pivot the column named **'company'** into two columns using the function **pivot()**:

```python
data2.pivot(index = 'date', columns = 'company', values = 'closing price')
```

We can write the previous code, omitting the names of the arguments for **short codes**, like this:

```python
data2.pivot('date', 'company','closing price')
```

**Pivoting** can be also done by using two functions which are **set_index()** and **unstack()**, like this:

```python
frame1 = data2.set_index(['date', 'company'])
frame1
```

```python
frame1.unstack('company')
```

These two functions can be applied in one single line of code, like this:

```python
data2.set_index(['date', 'company']).unstack('company')
```
