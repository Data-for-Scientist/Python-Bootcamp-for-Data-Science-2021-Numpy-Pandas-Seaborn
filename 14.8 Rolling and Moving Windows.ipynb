## 14.8 Rolling and Moving Windows

First we import pandas library:

```python
import pandas as pd
```

Lets read this time series from a csv file using the function **pd.read_csv()**:

```python
apple = pd.read_csv('data/apple.csv', index_col = 'Date', parse_dates = True)
apple.head()
```

Before using rolling for the close price, lets plot it first using the function **plot()**:


```python
apple['Close'].plot()
```

We can apply rolling for the close price column using the function **rolling()** with roling window equals three:

```python
apple.rolling(window = 3 ).mean()
```

We can plot this time series before and after applying the rolling like this:

```python
apple['Close'].plot()
apple['Close'].rolling(window = 3).mean().plot()
```

To see how increasing the size of the rolling window affect this time series, lets increase the rolling window to **10**:

```python
apple['Close'].plot()
apple['Close'].rolling(window = 10).mean().plot()
```

Lets have another example, consider this time series:

```python
Air = pd.read_csv('data/AirPassenger.csv', index_col = 'Month', parse_dates = True)
Air.head()
```

Before using rolling, lets plot it first using the function **plot()**:

```python
Air['Passengers'].plot(figsize = (8,5))
```

Now lets plot this time series before and after applying the rolling method, like this:

```python
Air['Passengers'].plot(figsize = (8,5))
Air['Passengers'].rolling(window = 10).mean().plot()
```
