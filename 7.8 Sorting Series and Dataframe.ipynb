## 7.8 Sorting Series and Dataframe

First we import numpy and pandas libraries:

```python
import numpy as np
import pandas as pd
```

Here we create a series using the function **pd.Series()**:

```python
ser1 = pd.Series(np.random.randint(50, size =10), index = list('jsgjhfsagb'))
ser1
```

We can sort the index of this series using the function **sort_index()**:

```python
ser1.sort_index()
```

But notice that the original series is not changes:

```python
ser1
```

To make the sorting permenant, we use the argument **inplace = True**:

```python
ser1.sort_index(inplace = True)
ser1
```

We can also sort the data of the series using the function **sort_values()**

```python
ser1.sort_values()
```

Here we read a dataframe from a csv file using the function **pd.read_csv()**:

```python
wage = pd.read_csv('data/wage.csv', index_col = 'year')
wage
```

We can sort the index of this dataframe using the function **sort_index()**

```python
wage.sort_index()
```

We can also sort the labels of the columns by using the argument **axis = 1**:

```python
wage.sort_index(axis =1)
```

For descending ordering of the index, we use the argument **ascending = False**:

```python
wage.sort_index(ascending = False)
```

We can also sort the dataframe by a column, using the function **sort_values**: 

```python
wage.sort_values(by = 'age')
```

We can even sort the dataframe by multiple columns, like this:

```python
wage.sort_values(by = ['age', 'wage'])
```
