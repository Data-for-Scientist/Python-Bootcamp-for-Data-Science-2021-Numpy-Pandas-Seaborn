## 8.2 Reading Data In Text Format-part 2

First we import pandas library:

```python
import pandas as pd
```

Lets read a dataframe from a csv file using the function **pd.read_csv()**:

```python
df4 = pd.read_csv('data/auto1.csv', index_col = 'name')
df4.head()
```

We can also set the index to be multiple columns:

```python
df5 = pd.read_csv('data/auto1.csv', index_col = ['cylinders', 'mpg'])
df5.head(20)
```

We can skip certain rows from the data file using the argument **skiprows**:

```python
df6 = pd.read_csv('data/auto1.csv',skiprows= [1,2,4])
df6.head()
```

Lets read this data frame which contains **missing data**:

```python
df7 = pd.read_csv('data/auto3.csv')
df7.head()
```

We can check for missing data using the function **isnull()**:

```python
pd.isnull(df7)
```

we can use the function **any()** with the **isnull()** function to return single boolean value for each column:

```python
pd.isnull(df7).any()
```
