## 9.4 Removing Duplicate Entries

First we import numpy and pandas libraries:

```python
import numpy as np
import pandas as pd
```

Here we read a dataframe with **duplicate rows** using the function **pd.read_csv()**:

```python
dt1 = pd.read_csv('data/ex1.csv')
dt1.head()
```

To check for duplicate entries we use the function **duplicated()**:

```python
dt1.duplicated()
```

We can add the function **any()** to the prevous code to return a single boolean value to check for duplicates in the entire dataframe:

```python
dt1.duplicated().any()
```

WE can delete the duplicate entries using the function **drop_duplicates()**:

```python
dt1.drop_duplicates()
```

To make the changes reflected in the original dataframe, we use the argument **inplace = True**:

```python
dt1.drop_duplicates(inplace =True)
dt1
```
