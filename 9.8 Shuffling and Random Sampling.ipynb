## 9.8 Shuffling and Random Sampling

First we import numpy and pandas library:

```python
import numpy as np
import pandas as pd
```

Here we create a simple series 10 integers using numpy function **random.randint()**:

```python
ser1 = pd.Series(np.random.randint(20, size =10))
ser1
```

We can randomly shuffle the values in this series using the function **sample()**:

```python
ser1.sample(frac=1)
```

The index can be sorted again after the shuffling using the function **reset_index()**:

```python
ser1 = ser1.sample(frac=1).reset_index(drop=True)
ser1
```

To take a random sample or a random subset of your dataset we change the argument **frac** into the desired fraction, here we choose 50% of the data: 

```python
sample1 = ser1.sample(frac= .5)
sample1
```

```python
sample1 = ser1.sample(frac =0.5).reset_index(drop=True)
sample1
```

Here we read a dataframe from a csv file using the function **pd.read_csv()**:

```python
data = pd.read_csv('data/ex3.csv')
data
```

To select a random sample with 20% of this dataframe, we set the fraction argument to be **frac = 0.2**: 

```python
sample2 = data.sample(frac =0.2)
sample2
```

Also here the index can be sorted using the function **reset_index()**:

```python
sample3 = data.sample(frac =0.2).reset_index(drop=True)
sample3
```

We can also select a random sample based on the number of rows rather than fraction, here we choose a sybset of 100 rows:

```python
sample4 = data.sample(n=100).reset_index(drop=True)
sample4
```
