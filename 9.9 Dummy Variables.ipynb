## 9.9 Dummy Variables

First we import pandas library:

```python
import pandas as pd
```

Here we read a dataframe form a csv file using the function **pd.read_csv()**:

```python
data = pd.read_csv('data/ex4.csv')
data.head()
```

To create dummy variables from the categorical variable (marital), we use pandas function **pd.get_dummies()**:

```python
pd.get_dummies(data['marital'])
```

Here we create dummy variables from the categorical variable (sex):

```python
gender_dummies = pd.get_dummies(data['sex'])
```

```python
gender_dummies
```

We can add the created dummy variables to the dataframe using the function **join()**:


```python
data_with_dummies = data.join(gender_dummies)
```

We can display the head of the new dataframe using the function **head()**:

```python
data_with_dummies.head()
```
